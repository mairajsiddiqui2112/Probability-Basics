#Question 2: In a class of 50 students:

20 like Mathematics (M)

15 like Science (S)

5 like both subjects

What is the probability that a student chosen at random likes Mathematics or Science?

  - Total Students, N = 50
     (M) = 20
     (S) = 15
     ( M ∩ S ) = 5
  - (M ∪ S) =  (M) + (S) - ( M ∩ S )
            = 20 + 15 - 5
            = 30
    P ( M ∪ S) = 30/50
               = 3/5
               =0.6
    '''
    probability of students likes Mathematics or Science is 3/5 or .6
    '''

# Question 5: The average height of 1000 students = 160 cm. A sample of 100 students shows an average height
= 158 cm.
Find the sampling error.

 - Population Mean = 160 cm
 - Sample Mean = 158 cm

  - Sampling error = |Sample Mean - Population Mean|
                   = |158-160|
                   = |-2|
                   = 2
                  
  - The Sampling error is 2 cm.

# Question 8: From a deck of 52 cards, two cards are drawn without replacement. What is the probability that
both are Aces?

 - Probability of drawing First Ace:-
 - N( drawing an Ace ) = 4
 - N = 52
 - P ( drawing First Ace ) = N( drawing an Ace ) / N
                          = 4/52
                          = 1/13


 - Probability of drawing Second Ace:-
 -  N( drawing an Ace ) = 3
 - N = 51
 - P ( drawing Second Ace ) = N( drawing an Ace ) / N
                          = 3/51
                          = 1/17


  - Probability of Both are Aces:-
  - Combined probability = Probability of drawing First Ace x Probability of drawing Second Ace
                         = 1/13 x 1/17
                         = 1/221
                         = 0.0045

  '''
  The probability that both cards drawn are Aces is 0.0045'''

# Question 9: A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the
probability that all are non-defective?

- Percentage of defective bulbs = 2 %
- Thus, P( defective bulbs ) = 0.2

- P( non defective bulbs ) = 1 - P( defective bulbs)
                           = 1 - 0.2
                           = 0.98


 - 5 bulbs are chosen at random,
 - n=5
 - Hence, P ( all 5 bulbs are non defective ) = P(non defective bulbs) ^{n}
                                             = 0.98^{5}
                                             = 0.9039 (approx)

 '''
 The probability that all 5 chosen bulbs are non defective is 0.9039 '''

# Question 10: Differentiate between discrete and continuous random variables with examples.

#Discrete Random Variable
 - A discrete variable is a variable that is taken from a distinct value.
 - These values are countable.
 - These variables have finite data.
 - Discrete variables are always available in whole numbers.
  - Examples:-
•	Number of workers in an office
•	Number of babies born each day.
•	Number of steps you take in a day.

#Continuous Random Variable
 - A continuous variable is a variable that is taken from value within the range.
 - The values are measurable.
 - These variables have infinite data.
 - Continuous Random variables are usually available in decimals.
 - Examples:-
•	Time taken by an employee to reach office.
•	The weight of newborn babies.
•	The distance you walk in a day.
